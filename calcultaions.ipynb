```python
import os
import pandas as pd

folder = 'balance_sheet'
balance_sheets = {}
errors = {}

if not os.path.isdir(folder):
    print(f"Directory not found: {folder}")
else:
    files = sorted([f for f in os.listdir(folder) if f.lower().endswith('.csv')])
    if not files:
        print(f"No CSV files found in {folder}")
    else:
        for fname in files:
            path = os.path.join(folder, fname)
            try:
                df_bs = pd.read_csv(path)
                sector = df_bs["sector"][0]
                industry = df_bs["industry"][0]
                Shares_Outstanding = df_bs["Shares Outstanding"][0]
                Current_Price = df_bs["Current Price"][0]
                df_bs = df_bs.drop("Shares Outstanding", axis=1)
                df_bs = df_bs.drop("Current Price", axis=1)
                df_bs = df_bs.drop("sector", axis=1)
                df_bs = df_bs.drop("industry", axis=1)
                df_bs = df_bs.T
                df_col = df_bs.iloc[0]
                df_bs.columns = df_col
                df_bs = df_bs[1:].reset_index()
                
                df_bs = df_bs.rename(columns={'index': 'date'}) 
                df_bs['sector'] = sector
                df_bs['industry'] = industry
                df_bs['Shares Outstanding'] = Shares_Outstanding
                df_bs['Current Price'] = Current_Price
                balance_sheets[fname] = df_bs
                print(f"Loaded {fname} -> shape: {df_bs.shape}")
            except Exception as e:
                errors[fname] = str(e)
                print(f"Failed to load {fname}: {e}")

if errors:
    print("Errors for some files:")
    for f, err in errors.items():
        print(f, ":", err)
```

```python
folder2 = 'income_statement'
income_statement = {}
errors = {}

if not os.path.isdir(folder2):
    print(f"Directory not found: {folder2}")
else:
    files = sorted([f for f in os.listdir(folder2) if f.lower().endswith('.csv')])
    if not files:
        print(f"No CSV files found in {folder2}")
    else:
        for fname in files:
            path = os.path.join(folder2, fname)
            try:
                df_is = pd.read_csv(path)
                df_is = df_is.T
                df_col = df_is.iloc[0]
                df_is.columns = df_col
                df_is = df_is[1:].reset_index()
                
                df_is = df_is.rename(columns={'index': 'date'}) 
                income_statement[fname] = df_is
                print(f"Loaded {fname} -> shape: {df_is.shape}")
            except Exception as e:
                errors[fname] = str(e)
                print(f"Failed to load {fname}: {e}")

if errors:
    print("Errors for some files:")
    for f, err in errors.items():
        print(f, ":", err)
```

```python
df_combined = pd.DataFrame()
for cname, df_bs in balance_sheets.items():
    df_bs['company'] = cname.split('.')[0]
    df_combined = pd.concat([df_combined, df_bs])
```

```python
df_combined_financial = pd.DataFrame()
for cname, df_is in income_statement.items():
    df_is['company'] = cname.split('.')[0]
    df_combined_financial = pd.concat([df_combined_financial, df_is])
```

```python
# Convert to datetime
df_combined['date'] = pd.to_datetime(df_combined['date'], errors='coerce')

# Now check again
print(df_combined['date'].dtypes)

# Convert to datetime
df_combined_financial['date'] = pd.to_datetime(df_combined_financial['date'], errors='coerce')

# Now check again
print(df_combined_financial['date'].dtypes)
```

```python
df_combined_financial.reset_index(drop=True, inplace=True)
df_combined.reset_index(drop=True, inplace=True)
```

```python
df_final = df_combined_financial.merge(df_combined, on=['company','date'], how='left')
```

```python
# --- 1. Basic Working Capital Metrics ---
df_final['Working_Capital'] = df_final['Current Assets'] - df_final['Current Liabilities']
df_final['Working_Capital_Ratio'] = df_final['Current Assets'] / df_final['Current Liabilities']
df_final['Cash_Ratio'] = df_final['Cash And Cash Equivalents'] / df_final['Current Liabilities']
```

```python
import numpy as np

# Market Cap from yfinance info
df_final['Market_Cap'] = df_final['Shares Outstanding'] * df_final['Current Price']

# Total Liabilities = Total Assets - Equity
df_final['Total_Liabilities'] = df_final['Total Assets'] - df_final['Common Stock Equity']

# Tobin's Q
df_final['Tobins_Q'] = (df_final['Market_Cap'] + df_final['Total_Liabilities']) / df_final['Total Assets']
```

```python
# # --- 2. Net Trade Cycle (NTC) ---
# # For Inventory Days, Receivable Days, Payable Days we need Sales and COGS (if available).
# # If not in your dataset, you’ll need to merge income statement data.
# # Placeholder calculations (replace with actual Sales/COGS columns when available):

# df_final['Inventory_Days'] = (df_final['Inventory'] / df_final.get('COGS', np.nan)) * 365
# df_final['Receivable_Days'] = (df_final['Accounts Receivable'] / df_final.get('Revenue', np.nan)) * 365
# df_final['Payable_Days'] = (df_final['Accounts Payable'] / df_final.get('COGS', np.nan)) * 365
```

```python
# --- 4. Control Variables ---
df_final['Liquidity'] = df_final['Working_Capital_Ratio']
df_final['Leverage'] = df_final['Total Debt'] / df_final['Common Stock Equity']
df_final['Size'] = np.log(pd.to_numeric(df_final['Total Assets']))
# df_final['Growth'] = df_final.get('Revenue Growth', np.nan)
```

```python
import pandas as pd

# Assume df_final has columns: company, year, Total Revenue
df_final['year'] = pd.to_datetime(df_final['date']).dt.year

# Sort by company and year to ensure correct order
df_final = df_final.sort_values(by=['company','year'])

# Replace 0 with NaN to avoid division by zero
df_final['Total Revenue'] = df_final['Total Revenue'].replace(0, np.nan)

# Calculate revenue growth by company
df_final['Revenue_Growth'] = df_final.groupby('company')['Total Revenue'].pct_change()

# If you want percentage format
df_final['Revenue_Growth_pct'] = df_final['Revenue_Growth'] * 100

print(df_final[['company','year','Total Revenue','Revenue_Growth','Revenue_Growth_pct']].head(10))
```

```python
import numpy as np

# Ensure numeric types
df_final['Inventory'] = pd.to_numeric(df_final['Inventory'], errors='coerce')
df_final['Accounts Receivable'] = pd.to_numeric(df_final['Accounts Receivable'], errors='coerce')
df_final['Accounts Payable'] = pd.to_numeric(df_final['Accounts Payable'], errors='coerce')
df_final['Cost Of Revenue'] = pd.to_numeric(df_final['Cost Of Revenue'], errors='coerce')
df_final['Total Revenue'] = pd.to_numeric(df_final['Total Revenue'], errors='coerce')

# --- Inventory Days ---
df_final['Inventory_Days'] = (df_final['Inventory'] / df_final['Cost Of Revenue'].replace(0, np.nan)) * 365

# --- Receivable Days ---
df_final['Receivable_Days'] = (df_final['Accounts Receivable'] / df_final['Total Revenue'].replace(0, np.nan)) * 365

# --- Payable Days ---
df_final['Payable_Days'] = (df_final['Accounts Payable'] / df_final['Cost Of Revenue'].replace(0, np.nan)) * 365

# --- Net Trade Cycle ---
df_final['NTC'] = df_final['Inventory_Days'] + df_final['Receivable_Days'] - df_final['Payable_Days']
```

```python


# --- 5. Efficiency Index (Composite Score) ---
# Normalize metrics (0–1 scale)
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

df_final['Norm_WC_Ratio'] = normalize(df_final['Working_Capital_Ratio'])
df_final['Norm_NTC'] = 1 - normalize(df_final['NTC'])  # inverse, lower NTC is better
df_final['Norm_Leverage'] = 1 - normalize(df_final['Leverage'])  # inverse, lower leverage is better
df_final['Norm_Cash_Ratio'] = normalize(df_final['Cash_Ratio'])

# Weighted average efficiency score
df_final['Efficiency_Index'] = (
    0.3 * df_final['Norm_WC_Ratio'] +
    0.3 * df_final['Norm_NTC'] +
    0.2 * df_final['Norm_Leverage'] +
    0.2 * df_final['Norm_Cash_Ratio']
)

# --- 6. Sector Ranking (if sector column available) ---
if 'sector' in df_final.columns:
    df_final['Sector_Rank'] = df_final.groupby('sector')['Efficiency_Index'].rank(ascending=False)

# --- Final Output ---
# print(df_final[['company','year','Working_Capital','Working_Capital_Ratio','NTC','ROE','Tobins_Q','Efficiency_Index']].head())
```
